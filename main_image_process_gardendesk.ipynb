```python
import os
import shutil

# 路径设置
source_folder = './GardenDesk'
output_folder = './GardenDesk_Renamed'

if not os.path.exists(source_folder):
    print(f"错误: 找不到源文件夹")
else:
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.tiff')
    files = [f for f in os.listdir(source_folder) if f.lower().endswith(extensions)]
    files.sort()

    print(f"检测到 {len(files)} 张图片。正在处理...")

    for index, filename in enumerate(files, start=1):
        ext = os.path.splitext(filename)[1]
        
        # 改为 0001.jpg 格式，完美支持 3000 张图
        new_name = f"{index:04d}{ext.lower()}" 
        
        old_path = os.path.join(source_folder, filename)
        new_path = os.path.join(output_folder, new_name)
        
        # 自动覆盖已存在的同名文件
        shutil.copy2(old_path, new_path)
        
        # 每处理 500 张打印一次进度，避免 3000 张图运行过程中太安静
        if index % 500 == 0:
            print(f"进度: {index}/{len(files)}")

    print(f"完成！所有图片已处理至: {output_folder}")
```

```python
import cv2
import os

# 1. 路径
source_folder = './GardenDesk'
output_folder = './GardenDesk_H1024'

# 2. 创建输出目录
os.makedirs(output_folder, exist_ok=True)

# 3. 获取图片
extensions = ('.jpg', '.jpeg', '.png', '.bmp')
files = [f for f in os.listdir(source_folder) if f.lower().endswith(extensions)]
files.sort()

# 固定高度
target_height = 1024

print(f"处理 {len(files)} 张图片，目标高度: {target_height}px...")

for idx, filename in enumerate(files):
    img_path = os.path.join(source_folder, filename)
    img = cv2.imread(img_path)

    if img is None:
        print(f"跳过 {filename}")
        continue

    h, w = img.shape[:2]

    # 等比例计算宽度
    aspect_ratio = w / h
    target_width = int(target_height * aspect_ratio)

    # resize
    resized_img = cv2.resize(
        img,
        (target_width, target_height),
        interpolation=cv2.INTER_AREA
    )

    # 统一重命名（推荐）
    new_name = f"{idx:04d}.jpg"
    save_path = os.path.join(output_folder, new_name)

    cv2.imwrite(save_path, resized_img, [int(cv2.IMWRITE_JPEG_QUALITY), 95])

print("\n完成！")
print(f"输出路径: {os.path.abspath(output_folder)}")
```
